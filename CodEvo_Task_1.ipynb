```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/winequalityN - winequalityN.csv')
df.head()
df = df.dropna()
```

```python
y = df['quality']
X = df.drop('quality', axis=1)
# tp = df['type']
# X = X.drop('type', axis=1)
```

```python
X = X.drop('type', axis = 1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred = linear_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
# Initialize and train the DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(random_state=42)
param_grid = {
    'max_depth': [10,15, 20,25, 30],
    'min_samples_split': [5,7,10],
    'min_samples_leaf': [4,6,8]
}
grid_search = GridSearchCV(dt_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
y_pred_dt = grid_search.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)
print(f"Mean Squared Error (Decision Tree): {mse_dt}")
print(f"R-squared (Decision Tree): {r2_dt}")
```

```python

from sklearn.ensemble import RandomForestRegressor

# Initialize and train the RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=10, random_state=42) # Start with a small number of trees
rf_regressor.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_regressor.predict(X_test)

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Mean Squared Error (Random Forest): {mse_rf}")
print(f"R-squared (Random Forest): {r2_rf}")


# Ensemble: Combine predictions from Linear Regression and Random Forest
# Simple averaging
y_pred_ensemble = (y_pred + y_pred_rf) / 2

# Evaluate the ensemble
mse_ensemble = mean_squared_error(y_test, y_pred_ensemble)
r2_ensemble = r2_score(y_test, y_pred_ensemble)

print(f"Mean Squared Error (Ensemble): {mse_ensemble}")
print(f"R-squared (Ensemble): {r2_ensemble}")
```
